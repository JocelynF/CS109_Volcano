## Explain the Problem

### Background

The ability to predict volcanic eruptions has been a major area of study within volcanology for centuries. This is because volcanic eruptions have been known to create major humanitarian crises by causing large-scale losses of property and lives. Large-scale volcanic eruptions have the ability to effect global climate by adding large amounts of chemicals such as SO2 and CO2 into the atmosphere. For these reasons, it is important to understand the inner workings of volcanoes and what sort of processes effect their eruptions rates. 

### Related Work

While researching this topic we came across a variety of papers which discussed the seasonality of volcanic eruptions. Many people have shown that single volcanoes or subsets of volcanoes erupt more frequently during certain times of years than others [e.g. Sottilli and Palladino 2014]. One particular paper looked at volcanoes globally [Mason et al. 2004], and showed that there is a correlation between volcanic eruptions and eruption month. Although the researchers demonstrate that the seasonal fluctuations in eruptions correlate to hydrological cycles in certain regions (i.e. changes in water levels), they did not show the relationship between volcanic activity and tidal gravitational forcings that induce surface deformations on the earth (solid earth tides).  In addition, their findings made us curious as to why differences in seasonal patterns existed between regions. 

### Motivation

We seek to understand how various cyclical forces (also known as tidal forces) affect the rate of volcanic occurrences as well as the features of volcanoes that affect their susceptibility to these naturally occurring forces. Tidal forces are caused by celestial bodies, specifically the sun and the moon, as well as changes in the Earth's tilt and rotation. The combination of these forces can lead to movement in the crust on the meter scale. By changing the force exerted a specific location, volcanic processes might be greatly effected. For instance, less pressure may lead to gases exsolving from a magma within the Earth causing it to erupt. Alternatively, decreasing pressure may allow for more faults to be created which magma can travel through and erupt. Various studies have shown that volcanic eruptions follow a seasonal fluctuating pattern and that they seem to coincide with the solid earth deformations that occur from gravitational forces that have the most influence on Earth. We  hypothesize that there is a correlation between solid earth tides and global volcanic eruptions and that the relationship causes differences in volcanic activity throughout various regions in the world. For that reason, we seek to improve our understanding of the role that tidal forces play in global tectonic activity in order to predict the incidence of future volcanic activity. 

### Questions

Our questions can be summed up as the following:

(1) Do volcanic eruptions exhibit seasonal patterns?

(2) How well can we predict the phase at which a volcano will erupt based on volcanic and eruptive features?

(3) What specific volcano attributes cause a volcano to be more susceptible to tidal forces? For instance, it is possible that a volcano that lies on the oceanic crust may be more susceptible to seasonal forces than one that lies on the continental crust.

(4) Which tidal force correlates the strongest with volcanic eruptions?

###Dealing with Dates
The next step is to make sure the data is in a form we can use. First we must convert the start and end dates into datetime objects. There is one eruption that is listed as falling on a leap day during a non leap year. To account for this, we just converted the date to the 28th instead of the 29th. The calculation of tidal forces in the next requires that our dates be in Julian days, so we also added a Julian Start Dates column. 

```python
#Convert dates --- 2 days are on 2/29, one of those days is not actually a leap year---raises error---changed date from 2/29 to 2/28
from datetime import datetime
conf_erup['Start Day'][6724] = 28
conf_erup['Start Date'] = cleanup.convert_dates(conf_erup, 'Start Year', 'Start Month', 'Start Day')
conf_erup['End Date'] = cleanup.convert_dates(conf_erup, 'End Year', 'End Month', 'End Day')
conf_erup['Julian Start Dates'] = cleanup.julian_date(conf_erup, 'Start Date')
```

Next, we will impute the eruption length. To impute eruptions that do not have a value, we want to ensure that, as much as possible, we can use historical averages from volcanoes to predict additional eruptions at that volcano with an eruption length of NAN.  For those volcanoes, we aggregated the mean and std and then selected a value from the standard normal distribution.  For other volcanoes, with an eruption length that could not be computed due to the lack of an end date or any historical data on that volcanoes performance, we used the overall average and standard deviation of all volcanoes.  As a result, some of the values fell below 0, so for those volcanoes, we made the eruption = 0.  This was not a perfect method and it did bias our overall mean and standard deviation (because our domain forced the eruption length to be greater than 0).

```python
#impute eruption length
conf_erup['Eruption Length'] = conf_erup['End Date']-conf_erup['Start Date']      
erup_days = [ i/ np.timedelta64(1, 'D') for i in conf_erup['Eruption Length'].values]
conf_erup['Eruption Length'] = erup_days
group_mean = conf_erup.groupby('Volcano Number').mean()
group_std = conf_erup.groupby('Volcano Number').std()

print "Original"
print "Eruption Mean:",conf_erup['Eruption Length'].mean()
print "Eruption Std:",conf_erup['Eruption Length'].std()
print "Out of Total, Number of Eruptions with uncertain eruption times:",\
conf_erup[conf_erup['Eruption Length'].isnull()].shape[0]
not_null_group = group_mean[group_mean['Eruption Length']>=0]
not_null_std = group_std[group_std['Eruption Length']>=0]

#For volcanoes with a std of 0, given them a std of 1
vol_zero_std = not_null_std[not_null_std['Eruption Length']==0].loc[:,'Eruption Length'].index.values
#print vol_zero_std
for i in vol_zero_std:
    not_null_std.loc[i:i,'Eruption Length'] = 1

##Now for volcanoes that have a mean and std, we will impute eruption lengths based on those values
change_idx = conf_erup[(conf_erup['Eruption Length'].isnull()) & \
                (conf_erup['Volcano Number'].isin(not_null_group.index.values))]['Eruption Length'].index.values

def zero_bound(val):
    if val < 0:
        return 0
    else:
        return int(val)
    
new_values = [zero_bound(np.random.normal(m,s)) for m,s in zip(not_null_group['Eruption Length'].values,\
                                                   not_null_std['Eruption Length'].values)]
for i,n in zip(change_idx,new_values):
    conf_erup.loc[i:i,'Eruption Length'] = n

#For volcanoes with native data, we used their average and std historically. 
print "Total Eruptions:",conf_erup.shape[0]
print "Out of Total, Number of Eruptions with uncertain eruption times:",\
conf_erup[conf_erup['Eruption Length'].isnull()].shape[0]

#For those volcanoes without native data, we will be forced to use global averages
mmean= conf_erup['Eruption Length'].mean()
sstd = conf_erup['Eruption Length'].std()
affected = conf_erup[conf_erup['Eruption Length'].isnull()]

new_values = [zero_bound(np.random.normal(mmean,sstd)) for i in xrange(affected.shape[0])]

affected_idx = affected.index.values

for i,n in zip(affected_idx,new_values):
    conf_erup.loc[i:i,'Eruption Length'] = n
    
print "After imputation, Number of Eruptions with uncertain eruption times:",\
conf_erup[conf_erup['Eruption Length'].isnull()].shape[0]
eruption_length = conf_erup['Eruption Length']
print
print "After"
print "Eruption Mean:",conf_erup['Eruption Length'].mean()
print "Eruption Std:",conf_erup['Eruption Length'].std()
#print conf_erup['Eruption Length']
```

### Calculating Phases

Since tidal phases are just sums of sine curves, we can calculate the phase at which an eruption occured. To do this, we arbitrarily set the maximum value within a cycle to 0 degrees, then calculated where the eruption occured based on the period of the cycle. The period and name of each cycle are as follows:

* Fortnightly (Mf): 13.66 Days
* Monthly (Mm): 27.57 Days
* Semiannual (Ssa): 182.5 Days
* Diurnal: 15 Days
* Semidiurnal: 30 Days

Originally the phases were such that eruptions occuring before the maximum of the tide were negative values, while eruptions occuring after the max were positive values. However, the rose plots we created required the values run from 0 to 360 so in the following cell the values have been converted. 

```python
# Convert values from [-180, 180 ] to [0, 360]
conf_erup.loc[conf_erup['Mf Phases']<0, 'Mf Phases'] = conf_erup['Mf Phases'] + 360.
conf_erup.loc[conf_erup['Mm Phases']<0, 'Mm Phases'] = conf_erup['Mm Phases'] + 360.
conf_erup.loc[conf_erup['Ssa Phases']<0, 'Ssa Phases'] = conf_erup['Ssa Phases'] + 360.
conf_erup.loc[conf_erup['Diurnal Phases']<0, 'Diurnal Phases'] = conf_erup['Diurnal Phases'] + 360.
conf_erup.loc[conf_erup['SemiDiurnal Phases']<0, 'SemiDiurnal Phases'] = conf_erup['SemiDiurnal Phases'] + 360.
```

### Features

Here are some of the features of interest that we will be exploring and why they are significant

#####VEI
Considered to be the Volcanic Explosivity Index, the VEI provides a measure of how explosive a volcanic eruption was.  Values range from 0 to 8, with 0 being the lowest and 8 being the highest.   Explosions from 0 - 2 can occur on a weekly basis, 3 - 4 can occur on a weekly to yearly basis, 5 on a decadal basis. Volcanoes with 6 and beyond are very rare and occur very intermittently. 
To put it in perspective, in the 19th century there were approximately 6 volcanoes with a VEI of 5, 2 volcanoes with a VEI of 6 and 1 volcano with a VEI of 7.  The VEI of 7 was considered the largest and deadliest volcano in recorded history.

#####Eruption Length
While the majority of volcanoes have eruption lengths that last less than a year, about 15% erupt over a year, with the longest lasting two decades or more.  An eruption is considered "finished" when a certain measure of decompression is attained within the volcano.

#####Elevation
Volcanoes have various elevation ranges (usually between 1,000m to 6,000m). Given that volcanoes are known for their height, there may be some significance between the height of the volcano and its other features. Hence, we felt it important enough to include in our analysis.

#####Dominant Rock Type
Volcanic rocks can form a variety of compositions due to a variety of factors such as how deep the magma chamber was, how long it was in a magma chamber, the amount of volatiles in the magma etc. It has been suggested that the release of pressure due to tidal forces may allow bubbles to grow in the magma chamber causing eruption. Therefore, rock types formed with more volatiles may be more susceptible to tidal forces.

######Tectonic Setting
Volcanism generally occurs at plate boundaries either at subduction zones (such as the Western US or Japan) where plates are coming together, or rift zones (such as the East African Rift or the East Pacific Rise) where plates are spreading apart. If volcanism does not occur at a plate boundary it is refered to as intraplate volcanism (such as Hawaii or Iceland). The tectonic setting greatly effects the style of volcanism. For instance, subduction zone volcanoes tend to be much more explosive due to the high quantity of volatiles in the magma. 

Another feature of tectonic setting is the type of crust. There are generally two types of crust: continental or oceanic. Oceanic crust is much thinner and slightly more dense than continental crust. This may effect the amount of faulting or fractures that can occur which magma travels through as it erupts.

######Region/Latitude and Longitude
We have eruption data from a variety of latitudes and longitudes. Generally speaking longitude does not say much about a volcano. However, the latitude of a volcano can bea proxy for the climate of a volcano. Some studies have suggested that the amount of ice on a volcano is more likely to effect whether or not it erupts. If our hypothesis is true, that tidal forces are more important than glacial cycles or seasonal weather patters, latitude should not have much of an effect on the cycling of eruptions.

## Statistics

Because our data is angular, it requires the use of directional statistics to interpret. Our stats.py module uses directional statistics in order to calculate the mean angle, mean vector (a measure of angular dispersion), circular standard deviation, and variance. The big question to test is whether all eruptions occur at various phases of tidal forces with equal probability or whether there is a phases where more volcanoes are erupting.  We used the Rayleigh Test for uniformity to determine whether or not the data follow a von Mises distribution (circular normal distribution). This test is only a sensible test when the alternative to uniformity is unimodal. If we believe that a decrease in pressure would cause volcanic eruptions, then it is reasonable to assume that the distribution may be unimodal with most eruptions occuring at the cyclic low. However, when looking at the results of the Rayleigh test for uniformity, it is important to consider that the alternative may be bimodal or some other distribution.

The Shapiro-Wilk test was what we wanted to use to determine if the data followed the normal distribution and it was seen as the best test because (1) it is easy to interpret (relies on a p-value) similar to the test above. The p-value would tell us whether we should accept the null hypothesis that the data came from a distribution that is normal or reject the null hypothesis and assume it came from a distribution that is not normal.  In addition, the Shapiro-Wilk test does not require us to produce an estimate of the mean and variance upfront, which makes it very agreeable to exploratory analysis, where we seek to uncover these parameters of interest.
